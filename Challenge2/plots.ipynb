```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Query 1

The data we are reading, corresponds to the number of employees for department-job during 2021 divided by quarters

```python
df_1=pd.read_csv('Challenge_2_1.csv')
```

```python
df_1.index=df_1['department']+'-'+df_1['job']
df_1.drop(columns=['department','job'],inplace=True)
```

```python
plt.figure(figsize=(10,100))
ax=sns.heatmap(df_1,cmap = "viridis")
ax.tick_params(labelsize=5) 
```

```python
df_1.describe()
```

With the help of this analysis and the heatmap, drawing conclusions becomes more straightforward. The data presented in the heatmap offers a clearer perspective compared to the raw query results. The insights gathered from the graphic are as follows
<li>In Q1, the role of Research and Development-Recruiting manager saw the highest number of employees.</li> 
<li>Moving into Q2, the most staff were in the roles of Support-Web designer I and Support Teacher.</li>
<li>By Q3, the dominant positions shifted to Engineering-Quality Engineer and Support-Programmer IV employees. </li>
<li>In Q4, Engineering-Web designer employees were the most prevalent.</li>

This trend can be explained by the following reasoning: During Q1, the organization was actively searching for talent, and it was crucial to have a strong recruiting team in place. In Q2, they started hiring support staff, possibly to meet increased demand. By Q3, there was a surge in Quality and support engineers, possibly due to deployment-related issues. Finally, in the last quarter of the year, the prevalence of web designers suggests improved product stability."



## Query 2

The data below shows the number of employees by deparment, which hired more employees than the mean of hired employees during 2021

```python
df_2=pd.read_csv('Challenge_2_2.csv')
```

```python
df_2
```

```python
plt.figure(figsize=(8,4))
sns.set(style='whitegrid', rc={"grid.linewidth": 0.1})
sns.set_context("paper", font_scale=0.9)
color = sns.color_palette("Set2", 6)

#plot
ax=sns.barplot(df_2,x='department',y='department_employee_count', width=0.5, linewidth = 0.7)
ax.bar_label(ax.containers[0], fontsize=8)

# Labels and clean up on the plot                                                                                                                                                                                                                                                                                              
plt.xticks(rotation=15)
ax.set_xlabel("Department",fontsize=10)
ax.set_ylabel("Employees",fontsize=10)    
ax.tick_params(labelsize=10)                                                           
plt.tight_layout()                                                                                                                 
sns.despine(left=True, bottom=True)                                                   
plt.savefig('test.pdf', bbox_inches='tight')  
```

The graph indicates that the company likely belongs to the tech industry, as evidenced by the notable presence of support and engineering employees.
